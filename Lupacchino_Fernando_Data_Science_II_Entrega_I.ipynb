```python
Satisfaccion_Vuelos['Rango']= None
Satisfaccion_Vuelos.loc[(Satisfaccion_Vuelos['Age'] >= 7) & (Satisfaccion_Vuelos['Age'] <= 11), 'Rango'] = 'Nios'
Satisfaccion_Vuelos.loc[(Satisfaccion_Vuelos['Age'] >= 12) & (Satisfaccion_Vuelos['Age'] <= 18), 'Rango'] = 'Adolescentes'
Satisfaccion_Vuelos.loc[(Satisfaccion_Vuelos['Age'] >= 19) & (Satisfaccion_Vuelos['Age'] <= 44), 'Rango'] = 'Adulto Joven'
Satisfaccion_Vuelos.loc[(Satisfaccion_Vuelos['Age'] >= 45) & (Satisfaccion_Vuelos['Age'] <= 59), 'Rango'] = 'Adulto Medio'
Satisfaccion_Vuelos.loc[(Satisfaccion_Vuelos['Age'] >= 60) & (Satisfaccion_Vuelos['Age'] <= 74), 'Rango'] = 'Adulto Mayor'
Satisfaccion_Vuelos.loc[Satisfaccion_Vuelos['Age'] >= 75, 'Rango'] = 'Anciano'

Satisfaccion_Vuelos[['Age', 'Rango']].head()
```

```python
Satisfaccion_Vuelos['satisfaction_numeric'] = Satisfaccion_Vuelos['satisfaction'].map({'neutral or dissatisfied': 0, 'satisfied': 1})


cols_servicio = [
    'Inflight wifi service', 'Inflight entertainment', 'On-board service',
    'Leg room service', 'Baggage handling', 'Checkin service',
    'Inflight service', 'Cleanliness', 'satisfaction_numeric'
]

corr = Satisfaccion_Vuelos[cols_servicio].corr()

sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlacin entre servicios y satisfaccin')
plt.show()
```
